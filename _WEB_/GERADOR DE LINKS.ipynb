# GERADOR DE LINKS

```python
import os
import pandas as pd
from datetime import datetime, date
```

```python
class Geradordelinks():
    
    def __init__(self):
        
        self.existe_arquivo = False
        self.arquivo_nome = 'dados.csv'
        self.dados = None
        self.categoria = None
        self.dicionario = None
        
        self.Arquivo_Existe()
        
        if self.existe_arquivo:
            self.Arquivo_Ler()


    def Arquivo_Ler(self):
        
        self.dados = pd.read_csv(self.arquivo_nome, encoding = 'utf-16', sep = '\t')
        self.Modifica_Dados()
        
        
    def Arquivo_Gravar(self):
        
        self.Modifica_Dados()        
        self.dados.to_csv(self.arquivo_nome, encoding = 'utf-16', sep = '\t', index = False)
        
        
    def Arquivo_Existe(self):
        
        self.existe_arquivo = os.path.exists(self.arquivo_nome)
        
        
    def Modifica_Dados(self):
        
        colunas = ['categoria', 'descrição']

        for coluna in colunas:
            self.dados[coluna] = self.dados[coluna].apply(lambda x: x.upper())
        
        self.dados['favorito'] = self.dados['favorito'].apply(lambda x: x.replace('*', '⭐'))
        self.dados['importância'] = self.dados['importância'].apply(lambda x: x.replace('*', '⚠️'))
        self.dados['prioridade'] = self.dados['prioridade'].apply(lambda x: x.replace('*', '🔥'))
        
        
    def Add_Registro(self):
        
        teste, registro = self.Gerar_Registro()
        
        if teste:
            newdados = pd.DataFrame(registro, index = [0])
            self.dados = pd.concat([self.dados, newdados])
            
            self.Modifica_Dados()
        
        
    def Gerar_Registro(self):
        
        if input('Deseja cadastrar um novo registro') in ['s', 'S']:
        
            categoria = input('Informe a categoria:')
            descricao = input('Informe a descrição:')
            link = input('Informe o link:')
            importancia = input('Informe a importância')
            prioridade = input('Informe a prioridade')
            favorito = input('Favoritos')

            registro = {}

            registro['categoria'] = categoria
            registro['descrição'] = descricao 
            registro['link'] = link
            registro['data'] = str(datetime.now().date())
            registro['importância'] = '*' * int(importancia)
            registro['prioridade'] = '*' * int(prioridade)
            registro['favorito'] = '*' * int(favorito)

            return True, registro
        else:
            return False, _

    
gl = Geradordelinks()
```

```python
def salvar(nome, texto):
    arq = open(f'{nome}.txt', 'w', encoding = 'UTF-16')
    arq.write(texto)
    arq.close()


dados = gl.dados

listacategoria = list(dados['categoria'].unique())

for cat in listacategoria:
    texto = f'\n### {cat}\n\n'
    
    filtro = dados[dados['categoria'] == cat]
    for ct, dc, lk, dt, ip, pr, fv in filtro.values:
        texto += f'[{dc}]({lk})\n{fv}\n{ip} {pr}\n{dt}\n\n'
        
    
    salvar(cat, texto)
        
print(texto)
```
